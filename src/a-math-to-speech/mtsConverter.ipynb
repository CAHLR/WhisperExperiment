```python
hints = []

with open('hintStrings.txt', 'r') as file:
    hints = file.readlines()
file.close()
print(len(hints))
```

```python
import re

only_math_hints = []
for hint in hints:
    strings_between_dollars = re.findall(r'\$\$(.*?)\$\$', hint.replace('\n', ''))

    if strings_between_dollars == -1: # no math in this hint
        only_math_hints.append(' ')
    else:
        strings_between_dollars
        only_math_hints.append('@'.join(strings_between_dollars))

print(only_math_hints)
```

```python
# write to conversion file
with open('onlyMathPre.txt', 'w') as file:
    for math_hint in only_math_hints:
        file.write(math_hint + '$')
file.close()
```

```python
with open('onlyMathPost.txt', 'r') as file:
    math = file.readline()
file.close()
print(math)

math = math.split('dollars')
math = [hint.split('at sign') for hint in math]
# dollar => new line
# at sign => new character
math
```

```python
len(math)
```

```python

import copy
hints_done = copy.deepcopy(hints)
for i, hint in enumerate(hints):
    # i is the id of the hint basically
#
    math_replacements = math[i]
    for replacement in math_replacements:
        hints_done[i]  = re.sub(r'\$\$.*?\$\$', replacement, hints_done[i], count=1)

hints_done
```
